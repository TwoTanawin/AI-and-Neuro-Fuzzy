```python
import matplotlib.pyplot as plt

# Define the data points
a1 = [14, 16]
a2 = [20, 10]
a3 = [18, 6]
a4 = [8, 4]
a5 = [6, 10]

# Extract x and y values for each point
x_values = [a1[0], a2[0], a3[0], a4[0], a5[0], a1[0]]  # Add a1[0] to close the polygon
y_values = [a1[1], a2[1], a3[1], a4[1], a5[1], a1[1]]  # Add a1[1] to close the polygon

# Create a polygonal line plot
plt.plot(x_values, y_values, marker='o', linestyle='-', label='Polygon Line')

# Add labels to the points
for i, (x, y) in enumerate(zip(x_values[:-1], y_values[:-1])):  # Exclude the last point to avoid overlapping labels
    plt.text(x, y, f'a{i+1} ({x},{y})')

# Add labels to the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add a legend
plt.legend()

plt.grid(True)

# Show the plot
plt.show()
```
