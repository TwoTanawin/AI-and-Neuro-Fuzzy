```python
import numpy as np
import matplotlib.pyplot as plt

# Define a range of x values from 0 to 3
x = np.linspace(0, 3, 100)  # 100 points between 0 and 3

# Calculate the corresponding sigmoid values
sigmoid_values = 1 / (1 + np.exp(-x))

# Create a plot
plt.plot(x, sigmoid_values, label='Sigmoid Function', color='b')
plt.xlabel('x')
plt.ylabel('Sigmoid(x)')
plt.title('Sigmoid Function')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a range of x values
x = np.linspace(0, 3, 100)  # 100 points between 0 and 3

# Calculate the corresponding y values using the function
y = (x**3)/3 - (3*x**2)/2 + 2*x + 5

# Create a plot
plt.plot(x, y, label='f(x) = (x^3)/3 - (3x^2)/2 + 2x + 5', color='b')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x)')
plt.grid(True)
plt.legend()
plt.show()
```
